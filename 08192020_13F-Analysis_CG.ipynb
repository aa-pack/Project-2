```python
berkshire_df = berkshire_df.set_index('Quarter Date')
```

```python
berkshire_df
```

```python
berkshire_df_group = berkshire_df.loc['2018-12-31']
```

```python
brkb = berkshire_df_group.pivot_table(berkshire_df_group, columns=['Sector'])
```

```python
brkb = brkb.rename(columns={'COMMUNICATIONS':'XLC', 'CONSUMER DISCRETIONARY':'XLY', 'CONSUMER STAPLES':'XLP', 'ENERGY':'XLE', 'FINANCE':'XLF', 'HEALTH CARE':'XLV', 'INFORMATION TECHNOLOGY':'XLK', 'MATERIALS':'XLB', 'REAL ESTATE':'XLRE', 'TRANSPORTS':'XLI', 'UTILITIES AND TELECOMMUNICATIONS':'XLU'})
```
