# Part 2: Data Preprocessing - weight_df

```python
import numpy as np
import pandas as pd
from datetime import datetime, timedelta
```

```python
df = pd.read_csv("best_quarterly_wgts.csv", 
                  index_col="Date", parse_dates=True, infer_datetime_format=True)
df
```

```python
df1= df.resample('QS').mean()
df1.head()
```

```python
df2= df1.resample('BQ').mean()
df2
```

```python
df2["total"] = df2[df.columns].sum(axis=1)
df2
```

```python
df3 = pd.DataFrame(index= df2.index, columns=df.columns)
df3
```

```python
for i in range(len(df2.index)):
    new_weight = df2.loc[df2.index[i],:'XLY'] * (100/ df2.loc[df2.index[i], 'total'])
    df3.loc[df2.index[i]] = new_weight
```

```python
df3
```

```python
df3.to_csv("./Resources/best_quarterly_weight_data.csv")
```
